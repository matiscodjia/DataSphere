```python
import numpy as np
from matplotlib import pyplot as plt
from scipy.interpolate import interp1d 
```

```python
x = np.linspace(0,10,10)
y =x**2
plt.scatter(x,y)
```

```python
f = interp1d(x,y,kind='linear')
new_x = np.linspace(0,10,30)
new_y = f(new_x)
plt.scatter(x,y)
plt.scatter(new_x,new_y)
```

## On va simuler un modèle cubique avec du bruit

```python
x = np.linspace(0,2,100)
y = 1/3*x**3 - 3/5 *x**2 +2 +np.random.randn(x.shape[0])/20
plt.scatter(x,y)
```

## On défini une fonction qui va permettre de régler les différents coefficients du modèle cubique

```python
def f(x,a,b,c,d):
    return a*x **3 +b*x**2 +c *x +d
```

## A l'aide de curve_fit on lance un algorithme qui va trouver les meilleurs coefficients qui vont à la courbe (avec la méthode des moindres carrés)

```python
from scipy.optimize import curve_fit
import scipy
params, param_cov = curve_fit(f,x,y)
```

```python
plt.scatter(x,y)
plt.plot(x,f(x,*params),c='green')
```

## Minimisation

```python
def f(x):
    return x**2 +15*np.sin(x)
```

```python
x = np.linspace(-10,10,100)
plt.plot(x,f(x))
```

```python
result = scipy.optimize.minimize(f,x0=-5).x
```

```python
plt.plot(x,f(x), lw=3,zorder=-1)
plt.scatter(result,f(result),s=100,c='r',zorder=1)
plt.scatter(-5,f(-5),s=200,marker='+',c='green',zorder=1)
plt.show()
```
