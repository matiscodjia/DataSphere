```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
x = np.linspace(0,50)
y = x**2
plt.plot(x,y)
```

## plot(X,Y, label= ..., lw(line width) = ..., ls(line style) = ..., c= ...)

```python
plt.plot(x, np.sin(y), c='r')
```

```python
plt.figure(figsize=(12,8))
plt.plot(x,y, label='quadratic')
plt.plot(x, np.cos(y), c='r', label='cosinus')
plt.title('Simple Plot')
plt.xlabel('axe x')
plt.ylabel('axe y')
plt.legend()
plt.savefig('figure.png')
```

### plt.subplot(lines, rows, position)

```python
plt.subplot(2,1,1)
plt.plot(x,y,c='r')
plt.subplot(2,1,2)
plt.plot(x,y,c='b')
```

## Orientée objet pour représenter une figure

```python
fig, ax = plt.subplots(2,1,sharex=True)
#ax est un tableau numpy à une dimension
ax[0].plot(x,y,c='r')
ax[1].plot(x,np.sin(x),c='b')
```

## Exercice 
### On doit générer à partir d'un dictionnaire de valeurs aléatoires à n expériences un graphique à n courbes

```python
def graphique(dataset):
    n = len(dataset)
    plt.figure(figsize=(12,8))
    for k, i  in zip(dataset.keys(), range(1,n+1)):
        plt.subplot(n,1,i)
        plt.plot(dataset[k])
        plt.title(k)
    plt.show()
        
```

```python
dataset = {'Expérience : {}'.format(i): np.random.randn(100) for i in range(10)}
```

```python
graphique(dataset)
```
