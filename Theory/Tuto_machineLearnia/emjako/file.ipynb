```python
listings = pd.read_csv("../Datasets/listing_extrait.csv")
```

```python
listings["longitude"].shape
```

```python
import matplotlib.pyplot as plt

# Supposons que listings soit un DataFrame Pandas contenant vos donnes
figure=plt.figure(figsize=(9,6))
plt.scatter(listings["longitude"], listings["latitude"], s=2, c=listings["availability_365"], cmap='viridis')
plt.axis('off')
plt.colorbar(label='Availability')
plt.show()
```

```python
import cartopy.crs as ccrs
from cartopy.io.img_tiles import OSM
```

```python
osm_tiles = OSM()
```

```python
plt.figure(figsize=(10,10))
ax = plt.axes(projection=osm_tiles.crs)
ax.set_extent([2.2,2.5,48.8,48.91],crs=ccrs.PlateCarree())
ax.add_image(osm_tiles,12)
plt.scatter(listings["longitude"], listings["latitude"], s=1, c=listings["availability_365"], cmap='viridis',transform=ccrs.PlateCarree(),alpha=.1)
```

```python
from cartopy.io.img_tiles import GoogleTiles
```

```python
plt.figure(figsize=(10,10))
google_tiles = GoogleTiles()
ax = plt.axes(projection=google_tiles.crs)
ax.set_extent([2.2,2.5,48.8,48.91],crs=ccrs.PlateCarree())
ax.add_image(google_tiles,12)
plt.scatter(listings["longitude"], listings["latitude"], s=1, c=listings["availability_365"],transform=ccrs.PlateCarree(),alpha=.1)
```
